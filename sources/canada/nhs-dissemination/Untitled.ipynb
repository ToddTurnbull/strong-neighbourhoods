```python
# problems with characters in the column names: create clean.csv with these stripped out
# (encodings tried: latin-1 iso-8859-1 cp1252 us-ascii utf-8)
with open("NHS_dissemination.csv") as nhs, open("clean.csv", "w") as clean:
    lines = nhs.readlines()
    for i, line in enumerate(lines):
        if i == 0: # fix the first line
            line = line.decode('utf-8', 'ignore').encode('utf-8')
        clean.write(line)
            
    
```
